### Matrix multiplication

Two main way of performing multiplication in neural networks and deep learning

1.Elementwise Multiplication

2.Matrix multiplication(dot product)

There are two main rules that a matrix multiplication needs to be satisfy:

1.The **Inner dimension** must match:
* `(3,2) @ (3,2)` won't work
* `(2,3) @ (3,2)` will work
* `(3,2) @ (2,3)` will work

2.The resulting matrix has the shape of **outer dimension**
* `(2,3) @ (3,2)` -> `(2,2)`
* `(3,2) @ (2,3)` -> `(3,3)`



```python
# element wise multiplication
tensor * tensor
```

```python
# matrix multiplication
torch.matmul(tensor,tensor)
```

You get error when we try to multiply same shape matrices
* `(3,2) @ (3,2)` --> 'ERROR'

What we can do is : Tranpose on of the matrix to get the inner dimension same
*`(3,2) @ (2,3)` --> `(3,3)`

```python
tensor_A = torch.tensor([[1,2],
                         [4,5],
                         [7,8]])

tensor_B = torch.tensor([[11,12],
                        [14,15],
                        [17,18]])
```

```python
# to fix the dimension issue problem we do
torch.matmul(tensor_A,tensor_B.T)
```

```python
tensor_B.T
# the rows became column and column became rows
```

## Reshaping , stacking , squeezing and unsqueezing tensors

* Reshaping - reshapes an input tensor to a defined shape

* View - return a view of an input tensor of certain shape but keep the same memory as its original tensor

* Stacking- combine multiple tensor on top of each other (vstack) or side by side(hstack)

* squeeze - removes all `1` dimensions from a tensor

* Unsqueeze - add `1` dimension to a
target tensor

* permute- return a view of the input with dimensions permuted(swapped) in a certain way

```python
# lets create a tensor
x = torch.arange(1.,10.)
x,x.shape
```

```python
# add an extra dimension
x_reshaped = torch.reshape(x,shape=(9,1))
x_reshaped , x_reshaped.shape
# to reshape the original tensor length must be equal to the multiplication of the reshaping param
# for example : if original tensor is (2,3) the we can reshape it to (1,6)  or (6,1) or (2,3) or any multiple of 6
```

```python
# Change the view
z = x.view(1,9)
z,z.shape
```

```python
# changing something in view z will affect or change the things in x also
z[0][0] = 5
z,x
```

```python
# stack tensors on top of each other
x_stacked = torch.stack([x,x,x,x],dim=0)
x_stacked
```

```python
# torch.sqeeze() - remove all the 1 dimension will be removed
x = torch.zeros(3,1,2,4)
x.size()
# [[[
#    [[0,0]],
#    [[0,0]]
# ]],[[[[0,0]],[[0,0]]]]]
y = torch.squeeze(x,0)
y.size()
```

```python
z = torch.unsqueeze(x,3)
z.size()
```

```python
## torch.permute - rearranges the dimension of a target tensor in a specified order
x_original =torch.rand(224,224,3)
# permute the original tensor
x_permuted = x_original.permute(2,0,1) # shifts the indexs 2-0,0->1,1->2
x_permuted.shape
x_permuted
```

## Pytorch tensors and Numpy
Numpy is a popular scientific python numerical computing torch.library
and because of this , Pytorch has unctionality to interact with it

* Data in Numpy,want in PyTorch tensor -> `torch.from_numpy(ndarray)`
* Pytorch tensor -> Numpy -> `torch.Tensor.numpy()`

```python
# Numpy array to tensor
array = np.arange(1.0,8.0)
tensor = torch.from_numpy(array).type(torch.float32)
# warning when converting numpy to pytorch , pytorch reflects numpys default data type of float64
array,tensor
```

```python
# Change the value in array
array = array + 1
array,tensor
```

```python
# tensor to numpy
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor,tensor.dtype,numpy_tensor,numpy_tensor.dtype
```

```python
# What happens if we change the tensor
tensor = tensor + 1
tensor,numpy_tensor
```

## PYTORCH REPRODUCIBILITY
In short how a neural network learns :

`start with random numbers -> tensor operations -> update random numbers to try and make them better representations of data -> again -> again->again...`

To reduce the randomness in neural networks and pytorch comes the concept of a **random seeed**.

Essentially what the random seed does is 'Flavour' the randomness

```python
import torch

# creating two random tensors
tensor_A = torch.rand(3,4)
tensor_B = torch.rand(3,4)

print(tensor_A)
print(tensor_B)
print(tensor_A == tensor_B)
```

```python
# Lets make some random but reproducible tensor

RANDOM_SEED = 69
torch.manual_seed (RANDOM_SEED)
random_C = torch.rand(3,4)

torch.manual_seed (RANDOM_SEED)
random_D = torch.rand(3,4)

print(random_C)
print(random_D)
print(random_C == random_D)
```
